```python
#import pytest
import torch
from mmdet.models.backbones import EfficientNet
import mmdet
from mmdet.models.backbones import ResNeSt
from mmdet.models.backbones.resnest import Bottleneck as BottleneckS
from mmdet.models.backbones.resnet import ResNet
from mmdet.datasets import LogMiniDet
```

```python
mmdet.__version__
```

```python
model = EfficientNet(arch='b3', out_indices=(0, 1, 2, 3, 4, 5, 6))
model.train()

imgs = torch.randn(1, 3, 1024, 1024)
feats = model(imgs)

for feat in feats:
    print(feat.shape)
print(model)
```

```python
model = ResNeSt(
        stem_channels=64,
        depth=50,
        radix=2,
        reduction_factor=4,
        avg_down_stride=True,
        num_stages=4,
        out_indices=(0, 1, 2, 3))
model.train()

imgs = torch.randn(1, 3, 1024, 1024)
feats = model(imgs)

for feat in feats:
    print(feat.shape)
```

```python
model = ResNet(depth=50, num_stages=4, out_indices=(0, 1, 2, 3))
model.train()

imgs = torch.randn(1, 3, 1024, 1024)
feats = model(imgs)

for feat in feats:
    print(feat.shape)

print(model)
```
